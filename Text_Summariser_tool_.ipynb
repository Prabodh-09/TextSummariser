```python
from transformers import pipeline
from textwrap import fill

# Step 1: Load the pre-trained summarizer model
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# Step 2: Define a function to summarize text
def summarize_text(text, max_length=130, min_length=30):
    summary = summarizer(text, max_length=max_length, min_length=min_length, do_sample=False)
    return summary[0]['summary_text']

# Step 3: Get user input
def main():

    print("=== TEXT SUMMARIZER TOOL ===\n")
    print("Enter your long article below (press Enter twice to finish input):\n")

    # Multiline input from user
    input_lines = []
    while True:
        line = input()
        if line == "":
            break
        input_lines.append(line)

    user_input = " ".join(input_lines)

    # Step 4: Generate and print the summary
    if len(user_input.strip()) == 0:
        print("❌ No text provided. Please try again.")
    else:
        print("\n--- GENERATED SUMMARY ---\n")
        summary = summarize_text(user_input)
        print(fill(summary, width=80))

# Run the script
if __name__ == "__main__":
    main()
```
